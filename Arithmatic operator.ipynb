### Addition +

```python
a = 20.5 #float
b = 34 #integer
a+b
```

```python
a = 20.5 #float
b = 34.5 #float
a+b
```

```python
a = 20 #integer
b = 4+2j #complex
a+b
#Here 20 & 4 are real part and for 'a' no imaginary part,so it is considered as 0 & 2j in 'b' are imaginary part.  
```

## True is Treated as '1' and False as '0'

```python
a = True #Boolean
b = True #Boolean
a+b
```

```python
a = False #Boolean
b = False #Boolean
a+b
```

```python
a = 21 #integer
b = True #Boolean
a+b
```

```python
a = 21 #integer
b = False #Boolean
a+b
```

```python
a = 21.20 #Float
b = True #Boolean
a+b
```

### In general adding numbers is called Addition but when we are adding strings together we should call it as **Concatenation**

```python
#concatenation
name1 = 'Hello' #string
name2 = 'World' #string
name1 + name2
```

```python
name1 = 21  #integer
name2 = 'World' #string
name1 + name2
```

### `Concatenation` is only possible in `Addition` and for the `strings` only.

```python
name1 = '21' #anything between the quotes is string
name2 = 'World' #string
name1 + name2
#string + string
```

### Division 

Here in this we have **3** types:
1. `Normal` division(/) - Gives `exact value` returns with`float`
2. `Floor` division(//) - Gives `Quotient` value returns with `Integer`
3. `Modulus` (%)        - Gives `Remainder`value returns with `Integer`

```python
a = 10
b = 4
```

```python
#Normal division
a/b
```

```python
#Floor division
a//b
```

```python
#Modulus division
a%b
```
